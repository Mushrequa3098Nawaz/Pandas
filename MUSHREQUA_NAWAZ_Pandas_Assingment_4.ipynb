```python
#  Q1.- Make a pivot table of top 10 bowlers(in terms of wickets) with over(1-20) showing no. of wickets.  [Dificult-]




wickets_taken = df2[~df2['player_dismissed'].isnull()] 
wickets_taken
top_bowlers = wickets_taken.groupby('bowler')['player_dismissed'].count().sort_values(ascending = False).index[0:10].tolist() 
x=wickets_taken[wickets_taken['bowler'].isin(top_bowlers)]
x.pivot_table(index = "bowler", columns = "over", values = "player_dismissed", aggfunc = "count") 
```

```python
# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]
def func(current_batsman):

    new=df2.merge(df,left_on='match_id',right_on='id')
    new['player_of_match']==current_batsman

    runs=df2.groupby('batsman').get_group(current_batsman)#we get a groupby object
    runs_scored=runs['batsman_runs'].sum()#.sort_values(ascending=False).head(5)
    print("total runs:",runs_scored)

    new=df2.merge(df,left_on='match_id',right_on='id')
    man_of_the_match=new[new['player_of_match']==current_batsman].shape[0]
    print("no of man of the match award:",man_of_the_match)

    no_of_boundaries=df2[(df2['batsman']==current_batsman) & (df2['batsman_runs']>=4)].shape[0]
    print("no of boudaries:",no_of_boundaries)

    innings=df2[df2['batsman']==current_batsman].shape[0]
    print("no of innings played:",innings)
func('V Kohli')    
```

```python
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```python
def func(venue):

    new=df2.merge(df,left_on='match_id',right_on='id')
    a=new[new['city']==venue]

    runs=a.groupby('batsman') #we get a groupby object
    top_batsman=runs['batsman_runs'].sum().sort_values(ascending=False).head(4)
    print("top 4 batsman of ",venue,"is:",top_batsman)
    
func('Pune')    
```
