## Q3. Make a function to find top bowler(in terms of most wickets) against any team. Take input as team.

```python
df2
```

```python

def topBowler(team) :
    mask = df2['batting_team'] == team
    data = df2[mask]
    mask2 = data['player_dismissed'].isnull()
    data2 = data[~mask2]
    print(data2.groupby('bowler').size().sort_values(ascending = False).head(1))

topBowler("Sunrisers Hyderabad")
```

## Q5. Write a function which will take a input as batsman and in output it gives total runs scored against each team. [Medium]




```python
df2= pd.read_csv(r"C:\Users\nawaz\deliveries.csv")
df2
def func(current_batsman):
    data=df2['batsman']==current_batsman
    runs=df2[data].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)  
    print("RUNS SCORED BY:",current_batsman)
    print(runs)
func('V Kohli')    
```
